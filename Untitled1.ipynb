```python
from transformers import GPT2Tokenizer
from datasets import Dataset

# Load your text file manually
with open("motivational_quotes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Remove empty lines and strip whitespace
texts = [line.strip() for line in lines if line.strip()]

# Convert to Hugging Face Dataset
dataset = Dataset.from_dict({"text": texts})

# Load tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token

# Tokenize
def tokenize_function(example):
    return tokenizer(example["text"], padding="max_length", truncation=True, max_length=64)

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])
```

```python
from transformers import GPT2LMHeadModel, Trainer, TrainingArguments
import os

# ✅ Disable wandb completely
os.environ["WANDB_DISABLED"] = "true"

# Load GPT-2 model
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Set training arguments
training_args = TrainingArguments(
    output_dir="./motivational_gpt2",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=50,
    save_total_limit=2,
    logging_steps=10,
    warmup_steps=10,
    fp16=True,
    logging_dir="./logs"
)

# Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
)

# 🚀 Start training
trainer.train()
```

```python
def tokenize_function(example):
    tokens = tokenizer(example["text"], padding="max_length", truncation=True, max_length=64)
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])
```

```python
prompt = "Believe in yourself because"

inputs = tokenizer(prompt, return_tensors="pt")

# Generate text
outputs = model.generate(
    **inputs,
    max_length=50,
    temperature=0.9,      # creativity level
    top_p=0.95,           # nucleus sampling
    top_k=50,             # randomness control
    do_sample=True,       # allow randomness
    num_return_sequences=1
)

# Show result
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
